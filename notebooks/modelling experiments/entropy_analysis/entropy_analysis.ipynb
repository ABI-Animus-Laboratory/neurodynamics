```python
import pickle
import numpy as np
import nest
import pandas as pd
from scripts import entropy, model
```

```python
categorised_neurons_path = '/hpc/mzhu843/modelling/nest/DATA/processed/model1/categorised_neurons_5.pkl'
results_path = '/hpc/mzhu843/modelling/nest/results/entropy_analysis/'
G_e_range = [0, 5, 10, 15, 20, 25, 30, 35, 40, 45, 50]
G_i_range = [0, -5, -10, -15, -20, -25, -30, -35, -40, -45, -50]
# G_e_range = [5]
# G_i_range = [-25]
```

```python
with open(categorised_neurons_path, 'rb') as file:
    categorised_neurons = pickle.load(file)
```

```python
col_names = ['G_e', 'G_i', 'Place Entropy', 'Int Entropy', 'Full Entropy']
df = pd.DataFrame(columns = col_names)
```

```python
for G_e in G_e_range:
    for G_i in G_i_range:
        print(f'G_e:{G_e}, G_i:{G_i}')
        row = {'G_e': G_e, 'G_i': G_i}
        network = model.Model1(categorised_neurons, G_e=G_e, G_i = G_i)
        network.simulate()
        place_spike_trains = network.spike_trains_pyr
        int_spike_trains = network.spike_trains_int
        all_spike_trains = np.concatenate((place_spike_trains, int_spike_trains))
        entropy_analysis1 = entropy.EntropyAnalysis(place_spike_trains)
        row['Place Entropy'] = round(entropy_analysis1.compute_entropy(), 1)
        entropy_analysis2 = entropy.EntropyAnalysis(int_spike_trains)
        row['Int Entropy'] = round(entropy_analysis2.compute_entropy(), 1)
        entropy_analysis3 = entropy.EntropyAnalysis(all_spike_trains)
        row['Full Entropy'] = round(entropy_analysis3.compute_entropy(), 1)
        df = pd.concat([df, pd.DataFrame([row])], ignore_index=True)
```

```python
df.to_csv(results_path + 'entropy_results.csv')
```
