```python
import numpy as np
import nest
import matplotlib.pyplot as plt
from params import pyr_hcamp_deco2012

E_L = -70.0
tau_m = 10

lif_neuron = nest.Create('iaf_psc_alpha', 1)
lif_neuron.set(pyr_hcamp_deco2012.params)
lif_neuron.set({'V_m':-58})
vm_monitor = nest.Create('multimeter')
vm_monitor.set(record_from=['V_m'])

nest.Connect(vm_monitor, lif_neuron)

nest.Simulate(300.0)

dmm = vm_monitor.get()
Vms = dmm['events']['V_m']
ts = dmm['events']['times']

fig = plt.figure(figsize=(10,4))
plt.plot(ts, Vms)
plt.ylabel('Voltage (mV)')
plt.xlabel('Time (ms)')

nest.ResetKernel()
```
