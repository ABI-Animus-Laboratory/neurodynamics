```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
neuron_attributes_path = '/hpc/mzhu843/modelling/nest/results/neuron categorisation/neuron_attributes.csv'
```

```python
neuron_attributes = pd.read_csv(neuron_attributes_path)

spike_count = neuron_attributes['Spike Count']
location_specificity = neuron_attributes['Location Specificity']
categories = neuron_attributes['category'].astype('category')
unique_categories = ['Interneuron', 'Place', 'Silent']

# Create a colormap with the same number of colors as unique categories
cmap = plt.cm.get_cmap('viridis', len(unique_categories))

# Scatter plot with color-coded categories
scatter = plt.scatter(location_specificity, spike_count, c=categories.cat.codes, cmap=cmap, s=20)

# Create legend entries with correct colors
legend_entries = [plt.Line2D([0], [0], marker='o', color='w', markerfacecolor=cmap(i), label=category)
                  for i, category in enumerate(unique_categories)]

# Add legend
plt.legend(handles=legend_entries, loc='best')

# Add labels and title
plt.xlabel('Location Specificity')
plt.ylabel('Spike Count')

# Adjust axis limits
plt.xlim(0, 1)
plt.ylim(0, 250)  # Adjust the limit based on your data range

# Show plot
plt.savefig('/hpc/mzhu843/modelling/nest/plots/neuron categorisation/neuron characterstics scatter')
```
