```python
import matplotlib.pyplot as plt
import numpy as np
import pickle
```

```python
#Get biological data for small model

all_calcium_traces = np.load('/hpc/mzhu843/modelling/nest/DATA/processed/spikes/con_S_5_p.npy')
all_spikes_obs = np.load('/hpc/mzhu843/modelling/nest/DATA/processed/spikes/S_5_p.npy')

selected_neurons = [33, 35, 70, 75, 157]
small_calc_obs = []
small_spikes_obs = []

for neuron_id in selected_neurons:
    small_calc_obs.append(all_calcium_traces[neuron_id -1])
    small_spikes_obs.append(all_spikes_obs[neuron_id -1])

plt.plot(small_calc_obs[1][:1000])
plt.plot(small_spikes_obs[1][:1000])

np.save('/hpc/mzhu843/modelling/nest/DATA/processed/model2/setup2/small_place_obs',small_calc_obs)
np.save('/hpc/mzhu843/modelling/nest/DATA/processed/model2/setup2/small_spikes_obs',small_spikes_obs)

```

```python
#Get biological data for full model

all_calcium_traces = np.load('/hpc/mzhu843/modelling/nest/DATA/processed/spikes/con_S_5_p.npy')
all_spikes_obs = np.load('/hpc/mzhu843/modelling/nest/DATA/processed/spikes/S_5_p.npy')
categorised_neurons_path = ('/hpc/mzhu843/modelling/nest/DATA/processed/model1/categorised_neurons_5.pkl')
with open(categorised_neurons_path, 'rb') as file:
    categorised_neurons = pickle.load(file)

place_cell_ids = categorised_neurons['Place'].keys()
interneuron_ids = categorised_neurons['Interneuron'].keys()

place_obs = []
spikes_obs = []

for neuron_id in place_cell_ids:
    place_obs.append(all_calcium_traces[int(neuron_id) -1])
    spikes_obs.append(all_spikes_obs[int(neuron_id) -1])

np.save('/hpc/mzhu843/modelling/nest/DATA/processed/model1/place_obs', place_obs)
np.save('/hpc/mzhu843/modelling/nest/DATA/processed/model1/spikes_obs', spikes_obs)

int_obs = []
int_spikes_obs = []

for neuron_id in interneuron_ids:
    int_obs.append(all_calcium_traces[int(neuron_id) -1])
    int_spikes_obs.append(all_spikes_obs[int(neuron_id) -1])

np.save('/hpc/mzhu843/modelling/nest/DATA/processed/model1/int_obs', int_obs)
np.save('/hpc/mzhu843/modelling/nest/DATA/processed/model1/int_spikes_obs', int_spikes_obs)
```
